```python
import sys
import math
import time 
# Auto-generated code below aims at helping you parse
# the standard input according to the problem statement.
timefor20=0
class Witchkitchen:
    """
    this is the class for all potential moves for the codingame autumn challenge.
    """
    

    def __init__(self):
        """
        this will initialise the class lists
        """
        self.inventory=[]
        self.cpturn_inventory=[]
        
        self.spelllist=[]
        self.brewlist=[]
        self.learnlist=[]
        self.futurepaths=[]
        self.nextfuturepaths=[]
        self.count=0
        self.learn_tog=-1
        self.bias=[0,0,0,0]


        self.rest_coef=5
        self.brew_coef=10
        self.work_towards_coef2=0.2
        self.work_towards_coef=0.4
        self.learn_goes=7


        self.end_time=0
        self.start_time=0
    
    def errorchecker(self, iteam):
        
        if iteam=="potions":
            print(self.brewlist)
        elif iteam=="spells":
            print(self.spelllist)
        elif iteam=="inventory":
            print(self.inventory)
        elif iteam=="states":
            print(self.futurepaths)
        elif iteam=="scores":
            print(self.weighted_scores)
            
    
    def reset (self):

        self.inventory=[]
        self.cpturn_inventory=[]
        
        self.spelllist=[]
        self.brewlist=[]
        self.futurepaths=[]
        self.nextfuturepaths=[]
        self.learnlist=[]
        self.learn_tog=-1


    
    def add_cast(self,input):
        """
        this will add any spells to the class at the start of each turn
        """
        if input[-1]==True:
            input[-1]=1
        elif input[-1]==False:
            input[-1]=0
        self.spelllist+=[(int(input[0]),int(input[2]),int(input[3]),int(input[4]),int(input[5]),int(input[-2]),input[-1])]
                         

        
    def turn_input(self,action_type,input):
        """
        sorts the turns data into the approtriate list   input=action_id, action_type, delta_0, delta_1, delta_2, delta_3, price, tome_index, tax_count, castable, repeatable
        """
    

        if action_type=="BREW":
            self.brewlist+=[(int(input[0]),int(input[2]),int(input[3]),int(input[4]),int(input[5]),int(input[6]),int(input[7]),int(input[8]))]  
        elif action_type=="CAST":
            self.add_cast(input)
        elif action_type=="LEARN":
            self.learnlist+=[(int(input[0]),int(input[2]),int(input[3]),int(input[4]),int(input[5]),int(input[7]),int(input[8]))]
            
        
    def intentory_input(self,input, player):
        """
        Creates inventories for human player and CPU player;   input=[inv_0, inv_1, inv_2, inv_3, score]
        """

        
        if player==0:
            self.inventory=(int(input[0]),int(input[1]), int(input[2]), int(input[3]),int(input[4]), 0)
        elif player==1:
            self.cp_inventory=(int(input[0]),int(input[1]), int(input[2]), int(input[3]), 0)
    
            
    def player_check_potion(self,potion, state):
        """    will work out if sufficient iteams in inventory for the potion to be made
        """
        
        inventory=state[0]                     
        if inventory[0]+potion[1]>=0 and inventory[1]+potion[2]>=0 and inventory[2]+potion[3]>=0 and inventory[3]+potion[4]>=0:
            ### returns true if player can do the potion
            return True
        else:
            return False
    
    def comp_check_potion(self,potion,state):
        """
        will work out if sufficient iteams in inventory for the potion to be made
        """
   #     inventory=state[.....]
        if inventory[0]+potion[1]>=0 and inventory[1]+potion[2]>=0 and inventory[2]+potion[3]>=0 and inventory[3]+potion[4]>=0:
            ### returns true if computer can do the potion.
            return True
        else:
            return False
        
    def player_check_spell(self,state):
        """
        will work out if sufficient iteams in inventory for the spell  to be cast
        """
        
        inventory=state[0]                     
        if inventory[0]>0 and inventory[1]>=0 and inventory[2]>=0 and inventory[3]>=0 and sum(inventory[:4])<10:
            ### returns true if player can do the potion
            return True
        else:
            return False
                             
    def spellinventory_update(self, spell, state):
        """
        update the inventory figures
        """ 
        inventory=state[0]                     
        new_inventory=(spell[1]+inventory[0],spell[2]+inventory[1],spell[3]+inventory[2],spell[4]+inventory[3],inventory[4])
        
        return (new_inventory,state[1],state[2],state[3])
        
                                 
    def brewinventory_update(self, brew ,state):
        """
        update the inventory figures after potion making
        """ 
        inventory=state[0]
        new_inventory=(brew[1]+inventory[0],brew[2]+inventory[1],brew[3]+inventory[2],brew[4]+inventory[3],(self.brew_coef*brew[5])+inventory[4])
        return (new_inventory,state[1],state[2],state[3])  
                                 
              
                                 
    def spell_remove(self,potion,state):
        """
        method to update spelllist with castable to 0 when used
        """
        spelllist=list(state[1])   #list function copies the state[1] to make alteration specific                      
        if potion[-1]==0:
            p_index=spelllist.index(potion)
    
            spelllist[p_index]=tuple(list(spelllist[p_index][:-2])+[0,0])
 
        return (state[0],spelllist,state[2],state[3])
                  
    def learn_toggle(self):

        if self.count<self.learn_goes:  #  remove this as I dont think it is needed.  I might be wrong - and self.inventory[0]>0:
            for tombs in self.learnlist:
                if tombs[5]==0:
               
                    self.learn_tog=tombs[0]

    def tomb_assesser(self):

        self.tomb_rating=[]

        tbias=sum([abs(i) for i in self.bias])

        for learn in self.learnlist:
            if tbias!=0:
                coefficent=((1-(self.bias[0]/tbias))*(learn[1]*0.5))+((1-(self.bias[1]/tbias))*(learn[2]*2))+((1-(self.bias[2]/tbias))*(learn[3]*3))+((1-(self.bias[3]/tbias))*(learn[4]*4))
            else:
                coefficent=(learn[1]*1)+(learn[2]*2)+(learn[3]*3)+(learn[4]*4)
            self.tomb_rating.append([learn[0],coefficent,sum([abs(i) for i in learn[1:-2]])])
        

    def ingredient_bias(self):

        self.bias=[0,0,0,0]
        for learn in self.spelllist:
            self.bias[0]+=learn[1]
            self.bias[1]+=learn[2]
            self.bias[2]+=learn[3]
            self.bias[3]+=learn[4]

    def learnchoice(self):

        choice=self.tomb_rating[:self.inventory[0]]
        h=-99

        for index, value in enumerate(choice):
            finallcoef=value[1]-(max(index,0.01)/3)
            elemts_adjust=0.8+(1*(1/value[2]))
            finallcoef*=elemts_adjust
            if finallcoef>h:
                h=finallcoef
                self.learn_tog=value[0]
            


    def simu_wait(self, state):
        
        refreshed=[]
        
        for spell in state[1]:
            newly_castable=list(spell[:-2])+[1]+[spell[-1]]
            refreshed.append(newly_castable)
        firstmove=state[3]    
        if firstmove==[]:
            firstmove=["REST"]
        invte=list(state[0])
        invte[4]-=(self.rest_coef*(1/self.simturn))

            
        return (tuple(invte),refreshed,state[2],firstmove)
                
        
        
    def future_states(self):
                   
        state=(self.inventory,self.spelllist,self.brewlist,[])                         
        self.futurepaths.append(state)
    
                             
    def add_future_states(self,state):
        
        self.nextfuturepaths.append(state)   
                             
    
    def swap_state(self):
                             
        self.futurepaths=self.nextfuturepaths
        self.nextfuturepaths=[]
    
    
    def cycle_states (self,state):
        
        output=[]                 
        no_viable_potion=1
        
        #first off run through the potion I could brew and see if I can, defintly do this if poss.
        if len(state[2])>0:
            for potion in state[2]:
                if self.player_check_potion(potion,state)==True:
                    
                    state2=self.brewinventory_update(potion,state)
                    no_viable_potion=0
                        
                    if state2[3]==[]:
                        state3=(state2[0],state2[1],state2[2],[potion[0]])
                    else:
                        state3=state2

                    output.append(state3)
                        
        if len(state[1])>0 and no_viable_potion==1:                          
            for spell in state[1]:
                state2=self.spellinventory_update(spell,state)
 
                if self.player_check_spell(state2)==True and spell[5]==1:

                    if spell[-1]==1 and 9==1:
                        repeatedlist=[state2]
                        repeating=1
                        while repeating==1:
           #                 print("REPEATING",len(repeatedlist), file=sys.stderr, flush=True)
                            newstate=self.spellinventory_update(spell,repeatedlist[-1])
                            if self.player_check_spell(newstate)==True:
                                repeatedlist.append(newstate)
                            else:
                                repeating=0

                        for index,newstate in enumerate(repeatedlist):
                            if index==0:
                                if newstate[3]==[]:
                                    state3=self.spell_remove(spell,(newstate[0],newstate[1],newstate[2],[spell[0]])) 
                                else:
                                    state3=self.spell_remove(spell,newstate)

                                if state3 not in output:
                                    output.append(state3)

                            else:
                                if newstate[3]==[]:
                                    state3=self.spell_remove(spell,(newstate[0],newstate[1],newstate[2],[spell[0],index+1])) 
                                else:
                                    state3=self.spell_remove(spell,newstate)

                                if state3 not in output:
                                    output.append(state3)
                                

                    else:

                        if state2[3]==[]:
                            state3=self.spell_remove(spell,(state2[0],state2[1],state2[2],[spell[0]])) 
                        else:
                            state3=self.spell_remove(spell,state2)


                        if state3 not in output:
                            output.append(state3)
                    
        if no_viable_potion==1:
            output.append(self.simu_wait(state))  #lastly the possibility of waiting
        return output



    def cheap_distance(self,player,potion,state):



        if player==0:            
            in_play_invent=list(state[0][:-1])
            
        if player==1:
            in_play_invent=list(state[4][:-1])

        zpotion=potion[1:-3]
        count=0
        print("...==",zpotion, file=sys.stderr, flush=True)   
        print("...==",in_play_invent, file=sys.stderr, flush=True)       

        for i,j in enumerate(zpotion):
            count+=(min((in_play_invent[i]+j),0)*(i+1))

        return count

                                 
    def potion_distance(self, player, potion, state):
        """
        calculates the number of basic steps required to afford the potion, using simple 1-1 conversion 
        as a analog to distance from potion
        """ 
        if player==0:            
            in_play_invent=list(state[0][:-1])
            
        if player==1:
            in_play_invent=list(state[4][:-1])
            
        zpotion=potion[1:-3]
        count=0
                       
        for i,j in enumerate(zpotion[:0:-1]):
            if i<3:
                i=i+1
                while j+in_play_invent[-i]<0:
                    in_play_invent[-i]+=1
                    in_play_invent[-(i+1)]-=1
                    count+=1

        while zpotion[0]+in_play_invent[0]<0:
            in_play_invent[0]+=1
            count+=1
        return count
    
                             
    def Simulate_go(self):
        """
        code to simulate all potential combinations of 3 potions
        """
                   
        self.future_states()
        self.simturn=1
        for i in range(4):
            self.simturn+=1
            for game_state in self.futurepaths:
                for i in self.cycle_states(game_state):
                    self.add_future_states(i)
            self.swap_state()
        
    def state_assessor(self):
        
        
        self.weighted_scores=[]
        for out_comes in self.futurepaths:
            countlist=[]
            score=(out_comes[0][4])
            
            for potions in out_comes[2]:
                countlist+=[[self.cheap_distance(0,potions,out_comes),potions[5]]]

       #     countlist.sort(key=lambda x:x[0])
    
      #      scorea=(25-countlist[0][1])*self.work_towards_coef*countlist[0][0]
      #      scoreb=(25-countlist[1][1])*self.work_towards_coef2*countlist[1][0]
            countlist2=[]
            for i in countlist:
                iscore=(22-i[1])*self.work_towards_coef*i[0]
                countlist2.append(iscore)
   #         print(scoreb, file=sys.stderr, flush=True)
  #          print(max(countlist2),countlist2, file=sys.stderr, flush=True)
   #         score-=scorea
   #         score-=scoreb

            self.weighted_scores.append([max(countlist2),out_comes[3][0]])

        ###   
        ###   here I shall probably include a posotive for waiting and reloading spells
        ###
        
        self.weighted_scores.sort(key=lambda x:x[0],reverse=True)



        
        if self.weighted_scores[0][0]==self.weighted_scores[1][0]:
            tie_no=1
            target=self.weighted_scores[0][0]
            while target==self.weighted_scores[tie_no][0]:
                tie_no+=1
                if tie_no==len(self.weighted_scores):
                    break
            occur=[]
           
            self.weighted_scores=self.weighted_scores[:tie_no]
            print("ASSES==",self.weighted_scores, file=sys.stderr, flush=True)
            for i in self.weighted_scores:
   
                if i[1]!="REST":
                    occur+=[i[1]]

            try:
                res = max(set(occur), key = occur.count)
                self.weighted_scores=[res]
            except:

                self.weighted_scores=["REST"]
            

            
    def move_selector(self):      
 
        
        if self.learn_tog!=-1:
            print("LEARN", self.learn_tog)
        else:
            print("MS==",self.weighted_scores, file=sys.stderr, flush=True)
            use_id=self.weighted_scores[0]


            try:
                repeated=use_id[1]-1
                print("REPEATING", file=sys.stderr, flush=True)
                task=""
                for cast in self.spelllist:
                    if cast[0]==use_id[0]:
                        task="CAST"
                        break
                    
                print(task,use_id[0],repeated)

            except:
                if use_id=="REST":
                    print(use_id)
                
                else:
                    task=""
                    for brew in self.brewlist:
                        if brew[0]==use_id:
                            task="BREW"
                            break
                    if task=="":
                        for cast in self.spelllist:
                            if cast[0]==use_id:
                                task="CAST"
                                break
                
                    print(task,use_id)


asshead=Witchkitchen()
count=0
while True:

    asshead.start_time = time.time()
    asshead.reset()
    action_count = int(input())  # the number of spells and recipes in play
    for i in range(action_count):
        a=input().split()
        asshead.turn_input(a[1],a)
#   print(int(action_id),action_type,int(delta_0),int(delta_1),int(delta_2),int(delta_3),int(price), int(tome_index),int(tax_count),castable,repeatable, file=sys.stderr, flush=True)
   
    for i in range(2):
         
        asshead.intentory_input(input().split(),i)

    # To debug: print("Debug messages...", file=sys.stderr, flush=True)


    # in the first league: BREW <id> | WAIT; later: BREW <id> | CAST <id> [<times>] | LEARN <id> | REST | WAIT
    
    asshead.learn_toggle()
    asshead.count+=1
    
    if asshead.learn_tog!=-1:
        asshead.tomb_assesser() 
        asshead.ingredient_bias()
        asshead.learnchoice()  
       
    else:    
        asshead.Simulate_go() 
        
        asshead.state_assessor()
        
        asshead.end_time =time.time()
        timefor20+=asshead.end_time-asshead.start_time
        
        print(f"The execution time is: {asshead.end_time-asshead.start_time}" , file=sys.stderr, flush=True)     
        print(timefor20 , file=sys.stderr, flush=True)     
    asshead.move_selector()
```
